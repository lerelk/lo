```python
import pandas as pd
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding_errors = 'ignore')
```

```python
df
```

```python
x = df[['PRICEEACH', 'SALES']]
x
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = "k-means++", random_state = 45)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10, 8))
plt.plot(range(1, 11), wcss)
plt.title("Elbow Method")
plt.xlabel("No. of clusters")
plt.ylabel("WCSS value")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', n_init = 'auto', random_state = 45)
```
